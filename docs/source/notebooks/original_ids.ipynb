## Get your data ready

For this notebook, we simulate data.

Notice that we use `j + 1` and set `include_id_reference_dict=True`.

```python
df = bpd.SimBipartite().simulate()
bdf = bpd.BipartiteDataFrame(i=df['i'], j=(df['j'] + 1), y=df['y'], t=df['t'], include_id_reference_dict=True).clean()
display(bdf)
```

## Original ids

Because we set `include_id_reference_dict=True`, the BipartiteDataFrame will track when contiguous ids change.

The method `.original_ids()` will return a dataframe that merges in the original ids.

```python
bdf.original_ids()
```
